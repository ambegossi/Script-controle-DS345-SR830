```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.loadtxt('xlist.txt')
```

```python
y = np.loadtxt('ylist.txt')
```

```python
freq = np.loadtxt('freqlist.txt')
```

```python
arq = open('parameters.txt', 'r')

params = arq.readlines()
for linha in params :
    print(linha)
arq.close()

# parameter units:
# freq = Hz / amplitude = V / sensitivity = V / time constant = s
```

```python
max = y.argmax()
plt.plot(freq,y)
plt.plot(freq[max],y[max],'x')
plt.annotate('freq = %f (Hz)'%(freq[max]), xy=(freq[max], y[max]))

plt.xlabel('Frequency (Hz)')
plt.ylabel('Y (V)')
plt.title("Cristal de Quartzo no ar")

```

```python
plt.figure()
max = x.argmax()
plt.plot(freq,x)
plt.plot(freq[max],x[max],'x')
plt.annotate('freq = %f (Hz)'%(freq[max]), xy=(freq[max], x[max]))

plt.xlabel('Frequency (Hz)')
plt.ylabel('X (V)')
plt.title("Cristal de Quartzo no ar")
```
